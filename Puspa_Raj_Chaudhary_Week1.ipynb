```python
# PART1

import numpy as np

# Define the matrices
A = np.array([[3, 4], [7, 8]])
B = np.array([[5, 3], [2, 1]])

# Task 1: Prove A.A^(-1) = I
A_inverse = np.linalg.inv(A)
identity_matrix = np.dot(A, A_inverse)  # A.A^(-1)
print("A.A^(-1):\n", identity_matrix)

is_identity = np.allclose(identity_matrix, np.eye(2))
print("Is A.A^(-1) equal to Identity Matrix (I)?", is_identity)

# Task 2: Prove AB != BA
AB = np.dot(A, B)
BA = np.dot(B, A)
print("\nAB:\n", AB)
print("BA:\n", BA)

is_commutative = np.array_equal(AB, BA)
print("Is AB equal to BA?", is_commutative)

# Task 3: Prove (AB)^T = B^T.A^T
AB_transpose = np.transpose(AB)
BT_AT = np.dot(B.T, A.T)  # B^T . A^T
print("\n(AB)^T:\n", AB_transpose)
print("B^T.A^T:\n", BT_AT)

is_associative = np.allclose(AB_transpose, BT_AT)
print("Is (AB)^T equal to B^T.A^T?", is_associative)
```

```python
# PART2

# Define matrices A and B for the system of equations
A = np.array([[2, -3, 1],
              [1, -1, 2],
              [3, 1, -1]])

B = np.array([-1, -3, 9])

# Solve using the inverse method: X = A^(-1) * B
A_inverse = np.linalg.inv(A)
X = np.dot(A_inverse, B)

print("Solution to the system of equations (x, y, z):\n", X)
```
